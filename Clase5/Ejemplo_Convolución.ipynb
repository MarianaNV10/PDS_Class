Las secuencias a tener en cuenta son:

$$x[n]=[0,5,\vec{-3},9,-4,-1,0,-8]$$

$$h[n]=[\vec{1},5,-2,-6,3,4]$$

Para el vector **x[n]** se tiene que $$N_{1}=-2, M_{1}=5$$
Por otro lado, para el vector **h[n]** se tiene que $$N_{2}=0, M_{2}=5$$
Se llega a que el límite en el cual la convolución es diferente a cero es: $$-2\leq{n}\leq{10}$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def grap_plots(x, y, title, xlabel, ylabel, pos, sub=True, fig=(10,10)):
    plt.figure(figsize=fig)
    if sub:
        for i in range(0,len(title)):
            plt.subplot(pos[i][0],pos[i][1],pos[i][2])
            if len(x) > 1:
                plt.stem(x[i],y[i])
            else:
                plt.stem(x,y[i])
    
            plt.title(title[i], fontsize=20)
            plt.xlabel(xlabel, fontsize=18)
            plt.ylabel(ylabel, fontsize=18)
            plt.grid()
```

```python
x = [0,5,-3,9,-4,-1,0,-8]
h = [1,5,-2,-6,3,4]
y = np.convolve(x,h) #Convolución
print(f'Resultado de la Convolución {y}')

n_x = np.arange(-2,6,1) # Vector de tiempo de x[n]
n_h = np.arange(0,6,1) # Vector de tiempo de h[n]
n_y = np.arange(-2,11,1) #Vector de tiempo de y[n]
```

```python
titles = ['x[n]', 'h[n]', 'y[n]']
y_axis = [x,h,y]
x_axis = [n_x,n_h,n_y]
pos = [(2,2,1),(2,2,2),(2,2,3)]

grap_plots(x_axis,y_axis,titles,'Samples','Amplitude',pos, fig=(20,15))
```
