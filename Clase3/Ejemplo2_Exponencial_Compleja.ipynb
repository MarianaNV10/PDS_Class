```python
import numpy as np
import matplotlib.pyplot as plt
import cmath
%matplotlib inline
```

```python
def grap_plots(x, y, title, xlabel, ylabel, pos, sub=True, fig=(10,10)):
    plt.figure(figsize=fig)
    if sub:
        for i in range(0,len(title)):
            plt.subplot(pos[i][0],pos[i][1],pos[i][2])
            plt.stem(x,y[i])
            plt.title(title[i], fontsize=20)
            plt.xlabel(xlabel, fontsize=18)
            plt.ylabel(ylabel, fontsize=18)
```

```python
n = np.arange(-11, 11, 1)
#Xn = 0.7**(n)
angle = 2.37
r = 0.85
Xn_real = (r**n)*np.cos(n*angle)  #Real Part
Xn_img = (r**n)*np.sin(n*angle)   #Imaginary Part

Xn_mag = np.sqrt(Xn_real**2+Xn_img**2)  #Magnitude
Xn_angle = np.arctan(Xn_img/Xn_real)    #Angle

y_axis = [Xn_real, Xn_img, Xn_mag, Xn_angle]
titles = ['Real Part X[n]', 'Imaginary Part X[n]', 'Magnitude X[n]','Angle X[n]']
pos = [(2,2,1),(2,2,2),(2,2,3),(2,2,4)]

grap_plots(n,y_axis,titles,'Sampes','Amplitude',pos,fig=(20,15))
```
