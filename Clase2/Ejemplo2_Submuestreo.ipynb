Se tiene que para muestrear bien una señal analógica se debe de cumplir el teorema de Nyquist
$$f_s \geq 2f_{max}$$
Y, si se muestrea mal es porque este teorema no se está cumpliendo a la hora del muestreo

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f1 = 3.0 #Frecuencia de Nyquist
f2 = 1.0
fs = 4 #Tasa de Nyquist

t_o = np.arange(0,2*np.pi, 0.01)
t = np.arange(0, 2*np.pi, 1/fs)
#sig_o = 3*np.sin(2*np.pi*f1*t)+np.cos(2*np.pi*f2*t)
#sig = 3*np.sin(2*np.pi*f1*t_m)+np.cos(2*np.pi*f2*t_m)

sig_o = 3*np.sin(2*np.pi*f1*t_o)
sig = 3*np.sin(2*np.pi*f1*t)

plt.figure(figsize=(20,10))
plt.plot(t_o,sig_o, label='Original', color='r')
plt.plot(t, sig, label='Aliasing', color='g')
plt.title('Analog Signal', fontsize=20)
plt.ylabel('Amplitude', fontsize=18)
plt.xlabel('Time', fontsize=18)
plt.legend()
plt.grid()
```

De la gráfica anterior, se puede observar que al muestrear con una tasa de Nyquist menor a la postulada en el teorema de Nyquist, la señal que se recupera es muy diferente a la original debido a que no se tomaron las suficientes muestras.
