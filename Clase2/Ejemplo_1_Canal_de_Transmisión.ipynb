```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

A continuación se simulará dos de los efectos (atenuación y distorsión) del canal de transmisión en la señal.

- En primera instancia, se presenta la señal de información a transmitir.

```python
t = np.arange(0,2*np.pi, 0.1)
sig = 5*np.cos(2*t)
plt.plot(t,sig)
plt.title('Analog signal', fontsize=20)
plt.ylabel('Amplitude', fontsize=18)
plt.xlabel('Time', fontsize=18)
plt.grid()
```

- Se usa la división para simular el comportamiento de la atenuación, es decir, la señal se divide entre cierta cantidad que hace que esta se atenúe.

```python
sig2 = sig/10
plt.title('Attenuated signal', fontsize=20)
plt.xlabel('Time', fontsize=18)
plt.ylabel('Amplitude', fontsize=18)
plt.plot(t,sig2)
plt.grid()
```

- El ruido se genera a partir de una función de la librería Numpy, luego este ruido se le suma a la señal atenuada que sería entregada al otro lado del canal de transmisión.

```python
noise = np.random.randn(int(len(sig2)))/8
sig3 = sig2 + noise

plt.title('Attenuated Sinal with Noise', fontsize=20)
plt.xlabel('Time', fontsize=18)
plt.ylabel('Amplitude', fontsize=18)
plt.plot(t,sig3)
plt.grid()
```
