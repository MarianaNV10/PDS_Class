```python
import matplotlib.pyplot as plt
import matplotlib.patches as paches

from scipy.signal import tf2zpk

%matplotlib inline
```

Se tiene la siguiente transformada z causal:
$$X[z]=\frac{-z^{2}-z}{z^{2}-5z+6}$$

```python
num = [-1,-1,0]
den = [1,-5,6]
```

```python
z,p,k = tf2zpk(num,den)
unit_c = plt.Circle((0,0),1, fill=False, color='darkcyan', ls='--')

fig, ax = plt.subplots()
ax.add_patch(unit_c)
plt.plot(p.real,p.imag,'x')
plt.plot(z.real,z.imag,'o')
plt.show()
```

De antemano, se sabe que la señal es causal y, del diagrama de polos y ceros se puede interpretar que no es estable ya que la ROC al ser mayor que $|z|>3$ no contiene al círculo unitario.
