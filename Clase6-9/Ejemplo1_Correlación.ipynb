```python
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

En este ejercicio se realizará la correlación a través de las dos definiciones, es decir, usando la propia definición de la correlación y su relación con la convolución. Para ello se harán dos arreglos llenándolos aleatoriamente teniendo en cuenta que la muestra n=0 se toma al inicio de ambos arreglos.

```python
lx = 7
lh = 4
x = []
h = []

for i in range(lx):
    x.append(random.randint(0,6))
    
for m in range(lh):
    h.append(random.randint(0,4))
    
print(f'Arreglo x[n]: {x}')
print(f'Arreglo h[n]: {h}')
```

```python
corr = np.correlate(x, h, mode='full')
t = np.arange(0, len(corr), 1)

h_inv = h[::-1] #Se invierte la lista
corr2 = np.convolve(x, h_inv, mode='full')
t2 = np.arange(0,len(corr2), 1)

plt.figure(figsize=(15,14))

plt.subplot(2,1,1)
plt.stem(t,corr)
plt.title('Correlación', fontsize=20)
plt.xlabel('Muestras', fontsize=18)
plt.ylabel('Amplitud', fontsize=18)
plt.grid()

plt.subplot(2,1,2)
plt.stem(t2,corr2)
plt.title('Correlación por convolución', fontsize=20)
plt.xlabel('Muestras', fontsize=18)
plt.ylabel('Amplitud', fontsize=18)
plt.grid()
```

Se puede observar que ambas gráficas son iguales, esto se debe a que al realizarla por la convolución se debe se entregar invertida la secuencia que se vaya a mover a lo largo de la otra.
