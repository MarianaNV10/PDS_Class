```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = np.arange(-40, 40, 0.1) # Continous Time
sig = 3*np.cos(0.1*np.pi*t) # Analog Signal 

plt.ylabel('Amplitude', fontsize=18)
plt.xlabel('Time', fontsize=18)
plt.title('Continous Signal', fontsize=20)
plt.plot(t,sig, label='Continous')
plt.grid()
```

```python
d_t = np.arange(-40,41,2)  #Discrete Time
d_sig = 3*np.cos(0.1*np.pi*d_t)

plt.figure(figsize=(20,5))
plt.stem(d_t, d_sig, linefmt='r', label='Discrete Signal')
plt.plot(t, sig, label='Continous Signal',)
plt.ylabel('Amplitud', fontsize=18)
plt.xlabel('Time', fontsize=18)
plt.grid()
plt.legend()
```
